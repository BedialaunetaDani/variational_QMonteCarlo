# Analysis of the algorithm implementations
In this notebook we perform several checks of the correct functioning of the walker's trajectories

```python
import numpy as np

import tests as tests
import lib as lib
```

First, we make a histogram in which we can see how the walker samples a Gaussian

```python
print("For a trial move variance of 10:")
tests.check_random_walker_1D(10)
print("For a trial move variance of 1:")
tests.check_random_walker_1D(1)
print("For a trial move variance of 0.01:")
tests.check_random_walker_1D(0.01)
```

We can see how in a big range of trial move variances the gaussian is sampled correctly. However, when the variance is too small we don't sample it correctly. 

Next, we can find an optimal trial move variance by finding that for which teh acceptance ratio is, on average, 0.5:

```python
dim = 1
tm_sigma_init = 1
tm_sigma_opt = lib.find_optimal_tm_sigma(tests.Gaussian, np.array([1]), dim, tm_sigma_init, tol = 0.05)
print("found optimal trial move: ",tm_sigma_opt)
tests.check_random_walker_1D(tm_sigma_opt)
```

And we see how the result is satisfactory and we no longer need to make blind guesses on the trial move.
